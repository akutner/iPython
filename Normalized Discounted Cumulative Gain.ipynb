```python



def gain(rel,i):
    return float(rel)/math.log(int(i+1),2)


#where ls is a list of entries
def dcg(ls):
    dcg = 0
    for i in range(0,len(ls)):
        dcg+=gain(mappings[ls[9]],i)
    return dcg


#where scores is ordered list of scores
def dcg_scores(scores):
    dcg = 0
    for i in range(1,len(scores)+1):
        dcg+=gain(scores[i-1],i)
    return dcg


IDCG = dict()
DCG = dict()

for key in topic_IDCG.keys():
    scores = topic_IDCG[key]
    DCG[key] = dcg_scores(scores)
    sort_scores = sorted(scores, reverse = True)
    IDCG[key] = dcg_scores(sort_scores)

print(DCG)
print(IDCG)

numpy.divide(list(DCG.values()),list(IDCG.values()))


```
